```python
manifest = {}
for file in pathlib.Path("../data").glob("*.parquet"):
    table_name = file.stem

    print(f"Loading {table_name}...")
    df = pd.read_parquet(file)
    manifest[table_name] = df.columns.tolist()
    display(df.describe())
    display(df.head())
    print()
```
